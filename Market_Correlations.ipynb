# Recent Returns on tickers

```python
tickers = ['^NDX', '^DJI', '^GSPC', 'DX-Y.NYB', 'GC=F',]
start = dt.datetime(2018, 1, 1)
end = dt.datetime(2022, 5, 20)

returns = pd.DataFrame()

for ticker in tickers:
    data = web.DataReader(ticker, 'yahoo', start, end)
    data = pd.DataFrame(data)
    data[ticker] = data['Close'].pct_change()
    
    if returns.empty:
        returns = data[[ticker]]
    else:
        returns = returns.join(data[[ticker]], how = 'outer')

returns = returns.dropna()
print(returns.tail())
```

```python
tickers = ['^NDX', '^DJI', '^GSPC', 'DX-Y.NYB', 'GC=F']

start = dt.datetime(2018, 1, 1)
end = dt.datetime(2022, 5, 20)

returns = pd.DataFrame()

for ticker in tickers:
    data = web.DataReader(ticker, 'yahoo', start, end)
    data = pd.DataFrame(data)
    data[ticker] = data['Close'].pct_change()
    
    if returns.empty:
        returns = data[[ticker]]
    else:
        returns = returns.join(data[[ticker]], how = 'outer')

returns = returns.dropna()

correlation = returns.corr()

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_xticks(np.arange(correlation.shape[0]) + 0.5, minor = False)
ax.set_yticks(np.arange(correlation.shape[1]) + 0.5, minor = False)
heatmap = ax.pcolor(correlation, cmap = plt.cm.RdYlGn)
fig.colorbar(heatmap)
heatmap.set_clim(-0.5, 1)

ax.set_xticklabels(correlation.index)
ax.set_yticklabels(correlation.columns)

fig = plt.gcf()
fig.set_size_inches(12, 10.5)

plt.show()
```
